```python
# importing necessary libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

## Looking Over the Data

```python
# 1 load in the data
df = pd.read_csv('mushroom_data.csv', sep=',')
df.head(10)
```

The dataset contains 23 columns of data describing thousands of mushrooms.
The types of data are mostly categorical. It is possible to visualize most of them with bar graphs.

## Plotting Bar Graphs

```python

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.countplot(x=df['Stalk Root'], palette='Set1', hue=df['Stalk Root'], legend=False, order=df['Stalk Root'].value_counts().index)
plt.title(f"Stalk Root barplot")
plt.xticks(rotation=30, fontsize=8)

plt.subplot(1, 2, 2)
wedge_sizes = df['Stalk Root'].value_counts(normalize=True)
pie_labels = wedge_sizes.index

plt.pie(wedge_sizes, labels=pie_labels, autopct='%1.1f%%')
plt.axis('equal')
plt.title(f"Stalk Root pieplot")

plt.subplots_adjust(left=0.05, bottom=0.05, right=0.95, top=0.95, wspace=0.3, hspace=0.3)
plt.show()
```

```python
# creating a function to generate bar plots
def barplot(df, column_name):
    sns.countplot(x=df[column_name], palette='Set1', hue=df[column_name], legend=False, order=df[column_name].value_counts().index)
    plt.title(f"{column_name} Value Counts")
    plt.xticks(rotation=30, fontsize=8)
    plt.xlabel(column_name, fontsize=12)

# creating a function to generate pie plots
def pieplot(df, column_name):
    wedge_sizes = df[column_name].value_counts(normalize=True)
    pie_labels = wedge_sizes.index
    plt.pie(wedge_sizes, labels=pie_labels, autopct='%1.1f%%')
    plt.axis('equal')
    plt.title(f"{column_name} Value Distribution")
```

```python
# 2-4 plot each column
# list of all column headers
columns= df.columns.to_list()
# creatining a for-loop to plot each column
for i in range(len(columns)):
    # print(columns[i])
    count_unique = len(df[columns[i]].unique())
    if count_unique <= 5:
        # if the column has 5 or fewer unique values, create a pie plot
        pieplot(df, columns[i])
    else:
        barplot(df, columns[i])
    plt.show()
    plt.clf()  # Clear the current figure to avoid overlap in plots
```
