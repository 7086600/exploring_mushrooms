```python
# importing necessary libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

## Looking Over the Data

```python
# 1 load in the data
df = pd.read_csv('mushroom_data.csv', sep=',')
df.head(10)
```

The dataset contains 23 columns of data describing thousands of mushrooms.
The types of data are mostly categorical. It is possible to visualize most of them with bar graphs.

## Plotting Bar Graphs

```python
# 2-4 plot each column
# list of all column headers
columns= df.columns.to_list()

# creatining a for-loop to plot each column
for i in range(len(columns)):
    # print(columns[i])
    sns.countplot(x=df[columns[i]], palette='Set1', hue=df[columns[i]], legend=False, order=df[columns[i]].value_counts().index) #8 add ordering
    # 6 # rotates the value labels slightly so they donâ€™t overlap, also slightly increases font size
    plt.xticks(rotation=30, fontsize=10)
    # increases the variable label font size slightly to increase readability
    plt.xlabel(columns[i], fontsize=12)
    # 7 add an informative title
    plt.title(f"{columns[i]} Value Counts", fontsize=14)
    plt.show()
    plt.clf()  # Clear the current figure to avoid overlap in plots
```
